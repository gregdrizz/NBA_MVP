```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
import datetime
import matplotlib.pyplot as plt

from time import sleep
```

```python
 # Create a beautifulSoup instance for parsing the html content
    
def get_initial_table(url):
    data = requests.get(url)
    data = data.content

    souper = BeautifulSoup(data, "html.parser")

    # Find all the html table's row which contains the keys of the table
    keysRowTag = souper.find("thead")

    # keysCellTags = keysRowTag.select("th.poptip.sort_default_asc.center")
    keysCellTags = keysRowTag.select("th.poptip")
    columns_names = [value.text for value in keysCellTags]

    keysColTag = souper.find("tbody")


    # get rows from table
    rows = souper.findAll('tr')[0:]
    rows_data = [[td.getText() for td in rows[i].findAll('td')]
                        for i in range(len(rows))]
    
    outputDf = pd.DataFrame(rows_data)

    outputDf.columns = columns_names[1:]
    
    return outputDf
    
    



```

```python
def clear_dataframe(inputDf):
    outputDf = inputDf.copy(deep = True)
    outputDf = outputDf.drop_duplicates(subset=['Player'])
    outputDf.dropna(inplace=True)
    outputDf['MP'] = pd.to_numeric(outputDf['MP'], downcast="float")
    outputDf['3P%'] = pd.to_numeric(outputDf['3P%'], downcast="float")
    outputDf = outputDf[outputDf['MP'] >= 15]
    outputDf = outputDf.reset_index()
    outputDf.drop("index", axis = 1, inplace = True)
    
    outputDf.index += 1

    return outputDf
```

```python
def add_years_column(inputDf, season_string_to_insert):

    outputDf = inputDf.copy(deep = True)
    lengthOfColumns = len(outputDf.index) + 1
    outputDf.insert(loc=0, column="Season", value = pd.Series([season_string_to_insert] * lengthOfColumns))
    outputDf['Season'] = pd.to_numeric(outputDf['Season'])

    return outputDf
```

```python
def create_table(url, season):
    df = get_initial_table(url)
    df = clear_dataframe(df)
    df = add_years_column(df, season)
    
    return df
```

```python
def convert_point_values_to_percentages(inputDf, columns_to_convert):

    outputDf = inputDf.copy(deep = True)

    for column_name in columns_to_convert:
        outputDf[column_name] = outputDf[column_name].apply(lambda x: func(x))

    return outputDf

def func(x):

    if x != '':
        result = "{:.2f}".format((float(x) * 100))
        return result
    return x


```

```python
def merge_tables(input_dfs_list):
    """
    """

    merged_df = pd.concat(input_dfs_list)
    merged_df = merged_df.reset_index()
    merged_df.drop("index", axis = 1, inplace = True)
    merged_df.index += 1

    return merged_df
```

```python
dfs = []

for i in range(2001, datetime.datetime.now().year):
    df = create_table(f"https://www.basketball-reference.com/leagues/NBA_{i}_per_game.html", str(i))
    dfs.append(df)
    df

full_table = merge_tables(dfs)

```

```python
mp_table = full_table.filter(['Season', 'MP'])
mp_avg = mp_table.groupby('Season')['MP'].mean()
mp_avg = mp_avg.to_frame()
mp_avg
```

```python
def clear_dataframe_mvps(inputDf):
    outputDf = inputDf.copy(deep = True)
    outputDf.dropna(inplace=True)
    outputDf['MP'] = pd.to_numeric(outputDf['MP'], downcast="float")
    outputDf['3P%'] = pd.to_numeric(outputDf['3P%'], downcast="float")
    outputDf = outputDf[outputDf.Lg == 'NBA']
    outputDf = outputDf.reset_index()
  
    outputDf.drop("index", axis = 1, inplace = True)
    
    outputDf.index += 1

    return outputDf
```

```python
mvps_table = get_initial_table('https://www.basketball-reference.com/awards/mvp.html')

mvps_table = clear_dataframe_mvps(mvps_table)

mvps_table.insert(0, 'Season', value =  2022 + -mvps_table.index)
mvps_table_relevant = mvps_table.head(21)


mvps_table_relevant
```

```python
# Merged MP table

mvps_mp_table_relevant = mvps_table_relevant.filter(['Season', 'MP'])
mvps_mp_table_relevant = mvps_mp_table_relevant.sort_values('Season', ascending=True)
mvps_mp_table_relevant = mvps_mp_table_relevant.rename(columns={'MP':'MVP MP'})

combined_mp = pd.merge(mvps_mp_table_relevant, mp_avg, on="Season")

combined_mp
```

```python
# Plot - MP Chart

combined_mp.plot(x="Season", y=["MP", "MVP MP"])
plt.locator_params(integer=True)
plt.show()
```

```python
threeP_table = full_table.filter(['Season', '3P%'])
threeP_table_avg = threeP_table.groupby('Season')['3P%'].mean()
threeP_table_avg = threeP_table_avg.to_frame()
threeP_table_avg
```

```python
mvps_threeP_table_relevant = mvps_table_relevant.filter(['Season', '3P%'])
mvps_threeP_table_relevant = mvps_threeP_table_relevant.sort_values('Season', ascending=True)
mvps_threeP_table_relevant = mvps_threeP_table_relevant.rename(columns={'3P%':'MVP 3P%'})

combined_threeP = pd.merge(mvps_threeP_table_relevant, threeP_table_avg, on="Season")

combined_threeP
```
